```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

df2022 = pd.read_excel("/content/2022_年間売上表.xlsx", sheet_name=0)
df2023 = pd.read_excel("/content/2023_年間売上表.xlsx", sheet_name=0)

df_all = pd.concat([df2022, df2023], ignore_index=True)

df_summary = df_all.groupby(["商品", "売上年"], as_index=False)["金額（千円）"].sum()

output_path = "/content/売上集計表.xlsx"
df_summary.to_excel(output_path, index=False)

wb = load_workbook(output_path)
ws = wb.active

fill = PatternFill(start_color="F2F2F2",
                   end_color="F2F2F2",
                   fill_type="solid")

for cell in ws[1]:
    cell.fill = fill

wb.save(output_path)
```
